## Example: Synthesis of a square wave via a Fourier Series

Consider the followin square wave 
$$ f(t) = sign(sin(w_0 t)),\quad -\infty <t<\infty$$

A periodic signal of period $T = \frac{\omega_0}{2\pi}$ can be represented
via a Fourier series 


$$ f(t) = \sum_{n=-\infty}^\infty F_n e^{i \omega_0 n t}$$

where the coefficients can be computed by solving the following integral


$$F_n = \frac{\omega_0}{2 \pi} \int_{t_0}^{t_0+T} f(t) e^{-i \omega_0 n t} dt$$


$$ f(t) \approx \sum_{n=-N}^N F_n e^{i \omega_0 n t}$$

$$ F_n = \frac{\omega_0}{2 \pi} \int_{-T/2}^{T/2} f(t) e^{-i \omega_0 n t} dt$$ 


```julia
using PyPlot
```

```julia
 L = 202
 M = 2*L + 1
 N = 21

 t = [(n) for n=-L:L]
 te1 = [(n) for n = -L-10:-10:-L-50]
 te2 = [(n) for n =  L+10: 10: L+50]
 T = 100
 f = zeros(Complex, M);
g = zeros(Complex, M);

F1 = zeros(Complex, N);
G = zeros(Complex, N);
F2 = zeros(Complex, N);
w0 = 2.0*pi/T
 
for k = 1:N
    F1[k] = (im/( pi*k))*(cos( pi*k)-1.0)
    F2[k] = (im/(-pi*k))*(cos(-pi*k)-1.0)
    G[k]= -(2.0/(pi*k))*(cos(pi*k)-1.0)
end


figure(1,figsize=(8,5))
subplot(211);
 plot(t,sign.(sin.(2*pi*t/T)),"-")
        plot(te1,zeros(length(te1)),".b")
        plot(te2,zeros(length(te2)),".b")
        grid("true")
xlabel(L"t [s]")
ylabel(L"f(t)")
title("Square wave")
arrow(0.0, 0.0,  T/2.0-5., 0., head_width=.1, head_length=4., linewidth=1, color="r")
arrow(0.0, 0.0, -T/2.0+5., 0., head_width=.1, head_length=4., linewidth=1, color="r")
text(1.3,-.29,L"T")

subplot(212)


for k = 1:N
for n = 1:M
    
        f[n] = f[n] + F1[k]*exp(im*w0*k*t[n]) + F2[k]*exp(im*w0*(-k)*t[n])
        g[n]= g[n]+ G[k]*sin(w0*k*t[n])
    end
    if k==21; plot(t,f,"-",label=L"N=21")
        plot(te1,zeros(length(te1)),".b")
        plot(te2,zeros(length(te2)),".b")
        grid("true");end
    if k==3; plot(t,f,"-",label=L"N=3")
        plot(te1,zeros(length(te1)),".b")
        plot(te2,zeros(length(te2)),".b")
        grid("true");end
     if k==1; plot(t,f,"-",label=L"N=1")
        plot(te1,zeros(length(te1)),".b")
        plot(te2,zeros(length(te2)),".b")
        grid("true");end
end

arrow(0.0, 0.0,  T/2.0-5., 0., head_width=.1, head_length=4., linewidth=1, color="r")
arrow(0.0, 0.0, -T/2.0+5., 0., head_width=.1, head_length=4., linewidth=1, color="r")
text(1.3,-.29,L"T")
xlabel(L"t [s]")
ylabel(L"f(t)")
title("Fourier synthesis" )
legend()
tight_layout()
savefig("Figs/square_wave.png")
```
